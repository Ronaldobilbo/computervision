```python

from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
# Download 'punkt_tab' data
nltk.download('punkt_tab')

sentences = [
    "Pengenalan pola adalah bagian penting dari machine learning.",
    "Machine learning membantu pengenalan pola secara otomatis.",
    "Model machine learning membutuhkan dataset yang besar.",
    "Pola dalam data membantu sistem pengenalan otomatis.",
    "Deep learning adalah cabang dari machine learning yang sangat penting."
]

tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

#Menampilkan vektor untuk kata tertentu(misalnya, "pola")
print("Vektor untuk kata 'pola':")
print(model.wv['pola'])  # Representasi vektor dari kata 'pola'

#Menampilkan kata-kata yang paling mirip dengan "machine"
print("\nKata yang mirip dengan 'machine':")
print(model.wv.most_similar('machine'))

#Menampilkan kata-kata mirip dengan learning
print("\nKata yang mirip dengan 'learning':")
print(model.wv.most_similar('learning'))
```

```python
import re
import pandas as pd
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

file_path = 'Gojekreview.csv'
data = pd.read_csv(file_path)

# Assuming the text data is in the 'Review' column
# Changed 'Gojekreview.csv' to 'Review'
texts = data['content'].tolist()  # Extract the 'Review' column and convert it into a list

# Simple tokenization using regular expressions
tokenized_texts = [re.findall(r'\b\w+\b', str(text).lower()) for text in texts]

# Train Word2Vec model
model = Word2Vec(sentences=tokenized_texts, vector_size=100, window=5, min_count=1, workers=4)

# Display the vector for a specific word, e.g., "mantap"
try:
    vector_mantap = model.wv['mantap']
    print("Vektor untuk kata 'mantap':")
    print(vector_mantap)
except KeyError:
    print("Kata 'mantap' tidak ditemukan dalam model.")

# Display words similar to "promo"
try:
    similar_to_promo = model.wv.most_similar('promo')
    print("\nKata-kata yang mirip dengan 'promo':")
    print(similar_to_promo)
except KeyError:
    print("Kata 'promo' tidak ditemukan dalam model.")

try:
    similar_to_bagus = model.wv.most_similar('bagus')
    print("\nKata-kata yang mirip dengan 'bagus':")
    print(similar_to_bagus)
except KeyError:
    print("Kata 'bagus' tidak ditemukan dalam model.")
```
